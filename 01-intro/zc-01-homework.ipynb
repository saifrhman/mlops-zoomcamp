```python
df['duration'] = round((df['tpep_dropoff_datetime'] - df['tpep_pickup_datetime']).dt.total_seconds() / 60,3)
df.head()
```

```python
df_feb['duration'] = round((df_feb['tpep_dropoff_datetime'] - df_feb['tpep_pickup_datetime']).dt.total_seconds() / 60,3)
df_feb.head()
```

```python
std_dev = round(df['duration'].std(),2)
print(std_dev)
```

```python
df_combined = pd.concat([df, df_feb], ignore_index=True)
df_combined.head()
```

```python
df_combined.shape
```

```python
df_filtered = df[(df['duration'] >= 1) & (df['duration'] <= 60)]
```

```python
df_filtered.head()
```

```python
outliers_num = df.shape[0] - df_filtered.shape[0]
records =  100 - ((outliers_num / df.shape[0] ) * 100)
records
```

```python
categorical = ['PULocationID', 'DOLocationID']
df_filtered[categorical] = df_filtered[categorical].astype(str)
df_filtered.head()
```

```python
df_filtered.shape
```

```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)

from sklearn.feature_extraction import DictVectorizer
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
```

```python
data_dicts = df_filtered[categorical].to_dict(orient='records')
vectorizer = DictVectorizer()
feature_matrix = vectorizer.fit_transform(data_dicts)
```

```python
print("Shape of the Feature Matrix:", feature_matrix.shape)
```

```python
model = LinearRegression()
model.fit(feature_matrix, df_filtered['duration'])
```

```python
predictions = model.predict(feature_matrix)
from sklearn.metrics import mean_squared_error
import numpy as np

rmse = np.sqrt(mean_squared_error(df_filtered['duration'], predictions))
print(f"RMSE: {rmse}")
```

```python
duration_february = (df_feb['tpep_dropoff_datetime'] - df_feb['tpep_pickup_datetime'])
df_feb['duration'] = duration_february.dt.total_seconds() / 60
filtered_data = df_feb.loc[(df_feb['duration'] > 1) & (df_feb['duration'] < 60)]
filtered_data
```

```python
filtered_data[categorical] = filtered_data[categorical].astype(str)
data_dicts = filtered_data[categorical].to_dict(orient='records')
feature_matrix = vectorizer.transform(data_dicts)
```

```python
feature_matrix.shape
```

```python
predictions = model.predict(feature_matrix)
```

```python
rmse = np.sqrt(mean_squared_error(filtered_data['duration'], predictions))
print(f"RMSE: {rmse}")
```
